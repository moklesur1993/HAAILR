```
from keras.layers import * #Dense, Dropout, Conv1D,MaxPooling1D,AveragePooling1D, Flatten
from keras.models import Sequential,Model,Input
from keras.layers.normalization import BatchNormalization
from keras.regularizers import l2
from keras.optimizers import *#RMSprop, Adam
from keras.callbacks import ReduceLROnPlateau , ModelCheckpoint,EarlyStopping
weight_decay = 1e-6
input_shape=[57,1]

model_input=Input(input_shape)
def build_discriminator(model_input):

    x = Conv1D(64, kernel_size=3,activation='relu',kernel_regularizer=l2(weight_decay))(model_input)
    x= BatchNormalization()(x)
    
    x = Conv1D(128, kernel_size=3,activation='relu')(x)
    x= BatchNormalization()(x)
    
    x = Conv1D(256, kernel_size=3,activation='relu',kernel_regularizer=l2(weight_decay))(x)
    x= BatchNormalization()(x)
            

    x=Flatten()(x)

    
    x=Dense(256, activation='relu',kernel_regularizer=l2(weight_decay))(x)# best for 32 and relu with out regularization & initiation
    x= BatchNormalization()(x)
    x=Dropout(.2)(x)
    x=Dense(128, activation='relu',kernel_regularizer=l2(weight_decay))(x)# best for 32 and relu with out regularization & initiation
    x= BatchNormalization()(x)
    out=Dense(12, activation='softmax')(x)
    model = Model(model_input, out)
    print("-- Discriminator -- ")
    model.summary()
    #model.compile(loss='binary_crossentropy', optimizer='adam')
    return model
  

first_model=build_discriminator(model_input)
def compile_and_train(model, num_epochs):
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy']) #Adamax for better
    checkPoint = ModelCheckpoint(filepath='/content/drive/My Drive/HAR/UCI_HAR/' + model.name +'.{epoch:02d}-{loss:.2f}.h5', verbose=1, save_best_only=True)
    #model=conv_pool_cnn_model
    earlystopping = EarlyStopping(monitor='val_acc',
                              patience=20,
                              verbose=1,
                              mode='auto')
    lr_reduce = ReduceLROnPlateau(monitor='val_loss', factor=.75, epsilon=0.001, patience=6, verbose=1) #0.5/.9

    history = model.fit(X_train,Y_train,batch_size=128,validation_data=(X_valid,Y_valid), shuffle=True,callbacks=[checkPoint,lr_reduce,earlystopping],epochs=80)#96&128

    #model.save_weights('/content/drive/My Drive/HAR/UCI_HAR/Test/aumentation.hdf5')
    return history
    

CNN1= compile_and_train(first_model, num_epochs=200)
vv=CNN1.model.evaluate(X_test,Y_test)
vv[1]
```
